```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#normal distribution

data = np.random.normal(loc=50,scale=10,size=100)
print(data)
plt.hist(data,bins=30)
plt.title("Normal Distribution")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

```python
# power law distribution
data1 = np.random.power(a=5,size=1000)
print(data1)
plt.hist(data1,bins=10,edgecolor='black')
plt.title('power distribution')
plt.xlabel("value")
plt.ylabel("freq")
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
data = pd.DataFrame({
    "age":np.random.randint(20,60,200),
    "salary":np.random.normal(50000,10000,200),
    "department":np.random.choice(["HR","IT","sales","finance"],200)
})
data
```

```python
#histogram of age
plt.hist(data['age'],bins=15,edgecolor="red")
plt.title("histogram of age distribution")
plt.xlabel("values")
plt.ylabel("frequency")
plt.show()
```

```python
#barplot (department counts)
dept_counts =data["department"].value_counts()
dept_counts
```
